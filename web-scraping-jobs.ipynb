```python
import requests 
import pandas as pd
import numpy as np  
import re
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import time
```

```python
url = 'https://djinni.co/jobs/?primary_keyword=Data+Science'
root_url = 'https://djinni.co/'

# define data for future dataframe
titles = []
urls = []
descriptions = []
locations = []
company_types = []
english_level_and_exp = []

page = 0
counter = 0
counter1 = 0

while True:
    # parse our html page
    r = requests.get(url, params = {"page" : page+1})
    html = r.content
    soup = BeautifulSoup(html, "lxml")
    # vacancies amount
    vacancies_count = int(soup.find(attrs={"class": "text-muted"}).text)
    # every page title
    title_page = soup.title.contents[0]
    print('{0}'.format(title_page))

    # titles
    for i in soup.findAll(attrs={"class": "profile"}):
        for j in i('span'):
            titles.append(j.text)

    # url of each vacancies
    for i in soup.findAll(attrs={"class": "list-jobs__title list__title order-1"}):
        for j in i('a'):
            urls.append(urljoin(root_url, j['href']))
            
    # short descriptions of vacancies
    for i in soup.findAll(attrs={"class": "list-jobs__description"}):
        for j in i('p'):
            descriptions.append(j.text)

    # job location
    for i in soup.findAll(attrs={"class": "location-text"}):
        sub = i.text.strip()
        sub = " ".join(sub.split())
        locations.append(sub)

    # type of companies, For example(remote, hybrid, office) and (product, outsorce, outstuff etc)
    for i in soup.findAll(attrs={"class": "list-jobs__details"}):
        company_types.append([])
        for j in i.findAll(attrs={"class": "ml-1"}):
            sub = j.text
            sub = re.sub(r'[^\w\s]', '', sub)
            sub = sub.replace("\n", "")
            sub = " ".join(sub.split())
            company_types[counter].append(sub)
        #print(company_types[counter])
        counter += 1

    # english level, and job experience 
    for i in soup.findAll(attrs={"class": "list-jobs__details"}):
        english_level_and_exp.append([])
        for j in i.findAll('nobr'):
            sub = j.text
            sub = re.sub(r'[^\w\s]', '', sub)
            sub = sub.replace("\n", "")
            sub = " ".join(sub.split())
            if sub not in company_types[counter1]:
                english_level_and_exp[counter1].append(sub)
        counter1 += 1
        
    # we can follow each link of our vacancies to work with individual job's page
    for f in urls:
        r = requests.get(f)
        html = r.content
        r_soup = BeautifulSoup(html, "lxml")
            
    page += 1
    
    if len(titles) == vacancies_count:
        break
len(urls)
```

```python
# define our columns with dictionary
columns = {'Title': titles, 'URL': urls, 'Location': locations, 'Type of job': company_types,
           'Leve of english, and job expirience': english_level_and_exp, 'Short discription of a company': descriptions}

# create dataframe
df = pd.DataFrame(columns)
df
```
